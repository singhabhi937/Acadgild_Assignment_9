```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")
```

```python
# Q1 delete the unnamed columns
```

```python
df.columns
```

```python
df[1:10]
```

```python
newdf=df.drop("Unnamed: 0",axis=1)
```

```python
newdf
```

```python
# Q2  Show the distribution of male and female
```

```python
newdf.groupby('Gender').count()
```

```python
# Q3 Show the top 5 most preferred names
```

```python
newdf.groupby('Name').count().sort_values(by='Id',ascending=False).iloc[0:5]
```

```python
# Q4. What is the median name occurence in the dataset
```

```python
newdf['Name'][np.median(newdf.index)]
```

```python
newdf['Name'].iloc[int(np.median(newdf.index))]
```

```python
# 5. Distribution of male and female born count by states
```

```python
newdf.groupby(['State','Gender']).count()['Id']
```
